```julia
using JLD2
using LinearAlgebra
using Plots
```

```julia
dict =load_object("Data/ohmic map data.jld2")

L_vec = dict["L_vec"]
Λ_vec = dict["Λ_vec"]
map_times = dict["map_times"]

spectra_L = zeros(ComplexF64,length(L_vec),size(L_vec[1])[1])
spectra_Λ = zeros(ComplexF64,length(Λ_vec),size(Λ_vec[1])[1])

[spectra_L[i,:] = eigen(L_vec[i]).values for i=1:length(L_vec)]
[spectra_Λ[i,:] = eigen(Λ_vec[i]).values for i=1:length(Λ_vec)]

plot(map_times,real.(spectra_Λ),xlabel="τD",title="Real part of map spectra",label=false)
```
